```python
cctv_seoul = pd.read_excel('./12_04_08_E_CCTV정보.xlsx')
cctv_seoul
```

```python
cctv_seoul_top1000 = cctv_seoul.head(1000)
cctv_seoul_top1000
```

```python
cctv_seoul_top1000 = cctv_seoul_top1000[['WGS84위도', 'WGS84경도']]
cctv_seoul_top1000
```

```python
import folium
import numpy as np
```

```python
import pandas as pd

cctv_seoul_top1000_df = pd.DataFrame(cctv_seoul_top1000)
cctv_seoul_top1000_df
```

```python
import pandas as pd
import folium
# 서울 지도 생성
map_seoul = folium.Map(location=[37.5665, 126.9780], zoom_start=12)

# DataFrame의 각 행을 순회하며 지도에 마커 추가
for index, row in cctv_seoul_top1000_df.iterrows():
    folium.Marker(location=[row['WGS84위도'], row['WGS84경도']]).add_to(map_seoul)

# 결과 저장 또는 출력
map_seoul.save('cctv_seoul_top1000.html')
```
