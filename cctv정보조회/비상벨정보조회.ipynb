```python
import folium

# 부산 지도 생성
map_busan = folium.Map(location=[35.1796, 129.0756], zoom_start=12)

# DataFrame의 각 행을 순회하며 지도에 마커 추가
for index, row in bell_busan.iterrows():
    folium.Marker(location=[row['WGS84위도'], row['WGS84경도']]).add_to(map_busan)

# 결과 저장 또는 출력
map_busan
```
