```python
cb = pd.read_csv('../ch09/CoffeeBean.csv', encoding='CP949', index_col=0, header=0)
cb
```

```python
addr = pd.DataFrame(cb.address.apply(lambda v: v.split()[:2]).tolist(), columns=['시도', '구군'])
addr
```

```python
addr.시도.unique()
```

```python
addr_aliases={'서울':'서울특별시', '서울시':'서울특별시'}
addr['시도']=addr['시도'].apply(lambda v: addr_aliases.get(v,v))
addr
```

```python
addr2 = addr.loc[addr.시도=='서울특별시',:]
addr2
```

```python
addr_group = addr2.groupby('구군')['시도'].count()

df = pd.DataFrame({'gugun':addr_group.index, 'count':addr_group.values})

df
```

```python
import requests
import json

r = requests.get('https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json')
c = r.content

seoul_geo = json.loads(c)
seoul_geo
```

```python
sido = open('../data/ctprvn.json', 'r', encoding='utf-8')
sido_geo = json.load(sido)
sido_geo
```

```python
m = folium.Map(
    location=[35.559819, 126.963895],
    zoom_start=11,
    tiles='cartodbpositron'
)

folium.GeoJson(
    seoul_geo,
    name='지역구'
).add_to(m)
m
```

```python
folium.Choropleth(
    geo_data=seoul_geo,
    data=df,
    columns=['gugun','count'],
    key_on='properties.name',
    fill_color='RdYlGn',
    bins=[0,1,2,3,4,5,10,20,30,57],
    fill_opacity=0.3,
    line_opacity=0.1,
    legend_name="지역구별 커피 업종 수",
    highlight=True).add_to(m)

m
```

```python
m = folium.Map(
    location=[37.559819, 126.963895],
    zoom_start=7,
    tiles='cartodbpositron'
)

folium.GeoJson(
    sido_geo,
    name='지역구'
).add_to(m)

m
```

```python
df2 = pd.read_excel('../data/행정구역_시군구_별__성별_인구수_20241223122518.xlsx', header=0)
df2
```

```python
folium.Choropleth(geo_data=sido_geo,
                    data=df2,
                    columns=['행정구역(시군구)별', '총인구수 (명)'],
                    fill_color='RdYlGn',
                    bins=[0,500000,1000000,2000000,5000000,90000000],
                    fill_opacity=0.3,
                    line_opacity=0.1,
                    key_on='properties.CTP_KOR_NM',
                    legend_name="시도별 총 인구수",
                    highlight=True).add_to(m)

m.save('./map.html')
```
