# Basic

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

```python
torch.manual_seed(1)
x_data=torch.FloatTensor([[1],[2],[3],[4],[5]])
t_data=torch.FloatTensor([[3],[5],[7],[9],[11]])
```

```python
model1=nn.Linear(1,1)       # input_dim=1, output_dim=1 이라는 뜻(y=wx+b)
print(list(model1.parameters()))
```

```python
optimizer=optim.SGD(model1.parameters(), lr=0.01)
epochs=20001
```

```python
for epoch in range(epochs):
    h=model1(x_data)
    cost=F.mse_loss(h, t_data)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

    if epoch%100 == 0:
        print('epoch: ', epoch, 'cost: ', cost.item())
```

```python
print(model1(torch.FloatTensor([[7],[8],[9]])))
print(list(model1.parameters()))
```
